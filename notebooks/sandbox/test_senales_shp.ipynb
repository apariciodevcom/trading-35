```python
# Validar senales para backtest
import pandas as pd
from pathlib import Path
from datetime import date

# Ruta del archivo de senales generado
CSV_PATH = Path("D:/trading/reports/senales_heuristicas") / f"senales_heuristicas_{date.today().isoformat()}.csv"

# Cargar y validar
if CSV_PATH.exists():
    df = pd.read_csv(CSV_PATH)
    columnas_esperadas = {"fecha", "simbolo", "estrategia", "signal"}
    if columnas_esperadas.issubset(df.columns):
        df["fecha"] = pd.to_datetime(df["fecha"], errors="coerce")
        df_valido = df.dropna(subset=["fecha", "simbolo", "estrategia", "signal"])
        print(f"Archivo listo para backtest: {CSV_PATH.name}")
        print(f"Total de filas validas: {len(df_valido)}")
        display(df_valido.head())
    else:
        print(f"Columnas faltantes. Se encontraron: {df.columns.tolist()}")
else:
    print("Archivo de senales no encontrado.")
```
