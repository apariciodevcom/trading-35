```python
import boto3
import os
from boto3.session import Session

# === Configuracion ===
# Ruta local deseada (relativa al notebook)
local_dir = os.path.abspath('../../data/historic')

# Crear directorio si no existe
os.makedirs(local_dir, exist_ok=True)

# Configurar boto3 con el perfil correcto
session = Session(profile_name='ses-trading')
s3 = session.client('s3')

# Configuracion S3
bucket_name = 'leantech-trading'
s3_prefix = 'data/historic_parquet/'

# Listar y descargar archivos
response = s3.list_objects_v2(Bucket=bucket_name, Prefix=s3_prefix)

if 'Contents' in response:
    for obj in response['Contents']:
        key = obj['Key']
        if key.endswith('.parquet'):
            file_name = os.path.basename(key)
            local_path = os.path.join(local_dir, file_name)
            print(f'Descargando {key} a {local_path}')
            s3.download_file(bucket_name, key, local_path)
else:
    print('No se encontraron archivos .parquet en el prefijo indicado.')

```
