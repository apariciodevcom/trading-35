```python
df_filtrado = df[
    (df['estrategia'].isin(['rsi_reversion', 'macd_hist_reversal', 'ma_envelope_reversals']))
]

ranking_simbolos = (
    df_filtrado
    .groupby('simbolo')
    .agg(
        total_trades=('retorno', 'count'),
        avg_retorno=('retorno', 'mean'),
        winrate=('retorno', lambda x: (x > 0).mean())
    )
    .sort_values(by='avg_retorno')
)

ranking_simbolos.head(10)  # Ver simbolos con peor rendimiento
```

```python
simbolos_malos = ranking_simbolos.query("avg_retorno < 0 or winrate < 0.45").index.tolist()
simbolos_malos
```
